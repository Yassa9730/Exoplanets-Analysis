```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr
```

```python
df = pd.read_csv('updated_exoplanets.csv')
```

```python
def analyze_discovery_trends(df):
    """Analyzes exoplanet discoveries over time and by detection method."""
    if 'discovery_year' in df.columns:
        plt.figure(figsize=(12, 6))
        df['discovery_year'].dropna().astype(int).value_counts().sort_index().plot(kind='bar', color='royalblue')
        plt.xlabel("Discovery Year")
        plt.ylabel("Number of Exoplanets Discovered")
        plt.title("Exoplanet Discoveries Over Time")
        plt.show()
    
        # Cumulative discovery plot
        df_sorted = df[['discovery_year']].dropna().astype(int).value_counts().sort_index().cumsum()
        df_sorted.plot(figsize=(12, 6), color='darkorange', marker='o')
        plt.xlabel("Year")
        plt.ylabel("Cumulative Discoveries")
        plt.title("Cumulative Growth of Exoplanet Discoveries")
        plt.show()
    
    if 'discovery_method' in df.columns:
        plt.figure(figsize=(10, 6))
        df['discovery_method'].dropna().value_counts().plot(kind='bar', color='mediumseagreen')
        plt.xlabel("Discovery Method")
        plt.ylabel("Number of Exoplanets")
        plt.title("Exoplanet Discoveries by Detection Method")
        plt.xticks(rotation=45, ha='right')
        plt.show()
```

```python
analyze_discovery_trends(df)
```
