```python
# Function to clean limit-flagged values
def clean_limit_values(df):
    limit_keywords = ['max', 'min', 'upper', 'lower', 'lim', 'ul', 'll', 'constraint', 'bound', '<', '>']
    
    # Identify columns that may contain limit values
    limited_cols = [col for col in df.columns if any(keyword in str(col).lower() for keyword in limit_keywords)]
    
    if limited_cols:
        print(f"Found {len(limited_cols)} columns possibly containing limit values.")

        for col in limited_cols:
            print(f"  - Cleaning column: {col}")

            # Replace string values like "<0.5" or ">10" with NaN, keep numeric values
            df[col] = df[col].astype(str).replace(r'[<>].*', np.nan, regex=True).astype(float)

    print("\nLimit-flagged values have been set to NaN where applicable.")
    return df
```

```python
# Function to identify outliers using IQR method
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Apply limit cleaning
df_clean = clean_limit_values(df)

# Get all numerical columns automatically
numerical_columns = df_clean.select_dtypes(include=['float64']).columns

# Identify outliers for each numeric column
for column in numerical_columns:
    if df_clean[column].notna().any():  # Check if column has non-null values
        outliers, lower_bound, upper_bound = find_outliers_iqr(df_clean, column)
        
        print(f"Outliers in {column}:")
        print(f"  Number of outliers: {len(outliers)}")
        print(f"  Lower bound: {lower_bound:.4f}")
        print(f"  Upper bound: {upper_bound:.4f}")
        
        # Create box plot for visualization
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df_clean[column])
        plt.title(f'Box Plot of {column}')
        plt.xlabel(column)
        plt.tight_layout()
        plt.show()
        
        # Display the first few outliers if any exist
        if len(outliers) > 0:
            print("  Sample outliers:")
            print(outliers[[column, 'planet_name', 'star_name']].head())
        
        print("\n" + "-"*50 + "\n")
```
